```python
#15-
import pandas as pd

df =  pd.read_csv('C:\\Users\\SHREE\\OneDrive\\Desktop\\Awadhut Study Material\\ecommerce_orders.csv')
# Replace missing Quantity values with 1
df['Quantity'] = df['Quantity'].fillna(1)

# Show updated missing values count to confirm
missing_values1 = df.isnull().sum()
print(missing_values1)
```

```python
#C-Matplotlib/seaborn Visualizations-18-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df =  pd.read_csv('C:\\Users\\SHREE\\OneDrive\\Desktop\\Awadhut Study Material\\ecommerce_orders.csv')
# Convert OrderDate to datetime format
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

# Group by date and sum TotalAmount
sales_over_time = df.groupby('OrderDate')['TotalAmount'].sum()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(sales_over_time.index, sales_over_time.values, marker='*', linestyle='-', linewidth=2, color='Red')
plt.title("Sales Over Time", fontsize=14, color='Green')
plt.xlabel("Order Date", fontsize=12)
plt.ylabel("Total Sales Amount", fontsize=12)
plt.grid(True, linestyle="-", alpha=0.6)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#19-
# Group by Region and sum TotalAmount
sales_by_region = df.groupby('Region')['TotalAmount'].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(8, 6))
sales_by_region.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Total Sales by Region", fontsize=14, color='Green')
plt.xlabel("Region", fontsize=12, color='red')
plt.ylabel("Total Sales Amount", fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#20-
# Group by Product and sum TotalAmount
product_sales = df.groupby('Product')['TotalAmount'].sum()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Product Sales Distribution", fontsize=14)
plt.tight_layout()
plt.show()
```
